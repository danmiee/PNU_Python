# 명언 크롤링

환경설정

```python
# 기본 모듈 임포트(re, usecsv, request, urllib.request > ur)
import re, usecsv
import urllib.request as ur
# BeautifulSoup > bs 임포트
from bs4 import BeautifulSoup as bs
```

웹사이트 정보 가져와서 자료형 변환하기

```python
# 웹사이트 정보 가져오기
url = 'https://quotes.toscrape.com/'
html = ur.urlopen(url)

# 뷰티풀수프로 자료형 변환
soup = bs(html.read(), 'html.parser')
```

```python
# 한 줄로 표현
soup = bs(ur.urlopen('https://quotes.toscrape.com/').read(),'html.parser')
```

태그에서 텍스트만 출력하기

```python
quote = soup.find_all('span')
for line in quote[:4]:
  # 태그 전체 가져오기
  print(line)
  # 클래스명 가져오기
  print(line.get('class'))
  # 텍스트 내용 가져오기
  print(line.text)
```

태그 안에 정의된 특정 클래스 찾아가기

```python
quote = soup.find_all('div', {"class" : "quote"})
```

정규표현식 활용하여 명언만 가져오기

```python
quote_res = [['명언', '인물명']]

for line in quote:
  # findall은 찾은 값을 리스트로 반환
  q = re.findall(r"[A-Z].+[a-z]\.", line.text)
  by = re.findall(r'by [A-Z][A-Za-z \.]+', line.text)
  # 리스트로 반환된 값을 더해주면 append 됨
  quote_res.append(q+by)
```

```python
# csv파일로 저장
usecsv.writecsv('quote.csv', quote_res)
```

# 포털 사이트 기사 크롤링

메인페이지에서 기사 제목 가져오기

```python
import re, usecsv
import urllib.request as ur
from bs4 import BeautifulSoup as bs

soup = bs(ur.urlopen('https://news.daum.net/').read(),'html.parser')

news = soup.find_all('a', {"class" : "link_txt"})

# news[-1].text

for line in news:
  print(line.text[37:-33])
  print(line.get('href'))
```

select를 이용하여 동명의 불필요한 태그 제거

```python
import re, usecsv
import urllib.request as ur
from bs4 import BeautifulSoup as bs

soup = bs(ur.urlopen('https://news.daum.net/').read(),'html.parser')

# find_all 보다 구체적으로 태그 선정
title = soup.select("div.box_g.box_news_issue > ul > li > div > div > strong > a")

for line in title:
  print(line.text.replace('  ','').replace('\n',''))
  print(line.get('href'))
  print()
```

제목 링크로 들어가서 기사 내용 가져오기

제목, 내용 함께 가져오기

```python
import re, usecsv
import urllib.request as ur
from bs4 import BeautifulSoup as bs

soup = bs(ur.urlopen('https://news.daum.net/').read(),'html.parser')
title = soup.select("div.box_g.box_news_issue > ul > li > div > div > strong > a")

for line in title:
  print(line.text.replace('  ','').replace('\n',''))
  print(line.get('href'))

  soup2 = bs(ur.urlopen(line.get('href')).read(),'html.parser')
  article = soup2.select("div.article_view > section > p")

  for line in article:
    print(line.text)
  print('\n')
```
